```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.collections as mc
from nltk.corpus import stopwords
```

```python
def get_random_colors(n):
    colors = []
    r = lambda: random.randint(0,255)
    for i in range(n):
        colors.append('#%02X%02X%02X' % (r(),r(),r()))
    return colors

def get_random_words(n, dictionary):
    r = lambda: random.randint(0,len(dictionary)-1)
    words = []
    for i in range(n):
        words.append(dictionary[r()])
    return words

def get_random_lines(n):
    starting_points = np.random.random(n)*100
    ending_points = np.random.random(n)*100
    return zip(starting_points, ending_points)

    
```

```python
def generate_figures(n):
    """generates random figures with different members, specify the number of figures you want"""
    for _ in range(n):
        #get the words and line coordinates
        stop = list(stopwords.words('english'))
        words = get_random_words(50,stop)
        lines = get_random_lines(10)

        fig = plt.figure()
        ax = fig.add_axes([0,0,1,1])
        ax.set_facecolor(get_random_colors(1)[0])
        x = np.linspace(0,1,100)
        y = np.linspace(1,10,100)


        # plot the curves
        for i in range(10):
            neg = np.random.random()
            a = np.random.random()
            b = 0
            if neg > 0.5:
                a = a * -1
                b = 100
            if np.random.random() > 0.5:
                plt.plot(x,a*y**2+b)
            else:
                plt.plot(x,a*y**(-2)+b)

        #plot the circles
        for i in range(45):
            point_x = np.random.random()
            point_y = random.randint(10,90)   
            plt.scatter(point_x, point_y,s=random.randint(100,500),color=get_random_colors(1)[0])

        #plot random lines
        for line in lines:
            plt.plot(line,c=get_random_colors(1)[0])
            plt.plot(np.array(line).T)

        # plot the words
        for word in words:
            ax.text(np.random.randint(10,90)/100., np.random.randint(10,90)/100., word,
                    horizontalalignment='left',
                    verticalalignment='top',
                    rotation=random.randint(0,180),
                    color=get_random_colors(1)[0],
                    fontsize=random.randint(5,20),
                    transform=ax.transAxes)

        plt.show()
```

```python
generate_figures(10)
```
