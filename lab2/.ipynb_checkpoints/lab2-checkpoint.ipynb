## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
```

## Load Datasets

```python
def load_iris():
    """load iris dataset
    Outputs:
    labels: 2d 1xn numpy array
    features: 2d 4xn numpy array
    """
    data = pd.read_csv('Fisher.csv')
    m = data.as_matrix()
    labels = m[:,0]
    labels[ labels == 2 ] = 1
    labels = np.atleast_2d( labels ).T
    features = m[:,1:5]

    return labels, features

def load_cifar():
    """load cifar dataset
    Outputs:
    labels: 2d 1xn numpy array
    features: 2d 3072xn numpy array
    """
    def unpickle( file ):
        import cPickle
        fo = open(file, 'rb')
        dict = cPickle.load(fo)
        fo.close()
        return dict
 
    data = unpickle( 'cifar-10-batches-py/data_batch_1' )

    features = data['data']
    labels = data['labels']
    labels = np.atleast_2d( labels ).T

    # squash classes 0-4 into class 0, and squash classes 5-9 into class 1
    labels[ labels < 5 ] = 0
    labels[ labels >= 5 ] = 1
    return labels, features
```

## Plotting Functions

```python
def plot_error(c,errors):
    for i in range(len(c)):
        plt.plot(errors[i], label="{}".format(c[i]))
    plt.xlabel("iterations")
    plt.ylabel("% accurate")
    plt.title("Classification Accuracy")
    plt.legend()
    plt.show()
    
def plot_norm(c,norms):
    for i in range(len(c)):
        plt.plot(norms[i], label="{}".format(c[i]))
    plt.xlabel("iterations")
    plt.ylabel("log l2 norm value")
    plt.title("Norm")
    plt.legend()
    plt.show()
```

## Perceptron Algorithm

```python
class Perceptron(object):
    def __init__(self, labels, features, c):
        self.features = features
        self.add_bias_to_features()
        self.weights = np.random.random(self.features.shape[1])
        #self.weights = np.zeros(self.features.shape[1])
        self.labels = labels
        self.c = c
        self.norms = []
        self.errors = []

    def get_weight_update(self, target, guess, x):
        # determines how far we need to step to get 'better' weights 
        delta_w = self.c * (target - guess)*x
        return delta_w
    
    def add_bias_to_features(self):
        #add bias to the features vector to homogenize the data
        b = np.ones((self.features.shape[0],1))
        self.features = np.concatenate((self.features,b),axis=1)
    
    def get_guess(self, x):
        output = self.weights.dot(x)
        if output > 0.0:
            return 1
        else:
            return 0
    
    def get_l2_norm_of_weights(self):
        #get the log of the 2 norm of the weights
        #norm = np.log(np.sqrt(self.weights.dot(self.weights)))
        return np.log(np.linalg.norm(self.weights,2))
    
    def get_error(self, targets, guesses):
        #this is actually accuracy (counting the number of correct guesses)
        correct = 0
        for i,t in enumerate(targets):
            if guesses[i] == t:
                correct += 1
        n = len(targets)
        return correct / float(n)
        
    
    def fit(self, n=100):
        #fit the perceptron to the data and collect the norms and accuracy(error)
        for _ in tqdm(range(n)):
            guesses = []
            for i,x in enumerate(self.features):
                guess = self.get_guess(x)
                guesses.append(guess)
                delta_w = self.get_weight_update(self.labels[i][0],guess, x)
                self.weights = self.weights + delta_w
            self.errors.append(self.get_error(self.labels, guesses))
            self.norms.append(self.get_l2_norm_of_weights())
                
```

```python
labels, features = load_iris()

c_iris = [1, 0.1, 0.01]
# c_iris = [0.01]

err = []
norm = []
for c in c_iris:
    p = Perceptron(labels, features, c)
    p.fit()
    err.append(p.errors)
    norm.append(p.norms)
    
plot_error(c_iris, err)
plot_norm(c_iris, norm)
```

```python
labels, features = load_cifar()

c_cifar = [0.001, 0.00001]

err = []
norm = []
for c in c_cifar:
    p = Perceptron(labels, features, c)
    p.fit()
    err.append(p.errors)
    norm.append(p.norms)
plot_error(c_cifar, err)
plot_norm(c_cifar, norm)
```
